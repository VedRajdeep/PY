```python
import pandas as pd
```

```python
x=pd.read_csv('car_evaluation.csv')#only data
```

```python
x.head()
```

```python
y=['buying','maint','doors','persons','lug_boot','safty','cars']#only headings
```

```python
x.columns=y
```

```python
x.head()
```

```python
x.info()
```

```python
buying={'vhigh':30,'high':20,'med':10,'low':5}
x.buying=[buying[i] for i in x.buying] 

maint={'vhigh':200,'high':150,'med':100,'low':50}
x.maint=[maint[j] for j in x.maint]

doors={'2':2,'3':3,'4':4,'5more':6}
x.doors=[doors[k] for k in x.doors]

persons={'2':2,'4':4,'more':5}
x.persons=[persons[l] for l in x.persons]

lug_boot={'big':1200,'med':800,'small':500}
x.lug_boot=[lug_boot[m] for m in x.lug_boot]

safty={'high':5,'med':4,'low':2}
x.safty=[safty[n] for n in x.safty]

cars={'vgood':1,'good':3,'acc':3,'unacc':4}
x.cars=[cars[o] for o in x.cars]
```
