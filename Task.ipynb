```python
city_complaint_counts = any.groupby(['City', 'Complaint Type']).size().unstack(fill_value=0)
```

```python
cities = any['City'].unique()
```

```python
fig, axes = plt.subplots(nrows=len(cities), figsize=(12, 6*len(cities)), sharey=True)
plt.subplots_adjust(hspace=0.5)

for i, city in enumerate(cities):
    city_data = city_complaint_counts.loc[city]
    city_data = city_data[city_data > 0]
    city_data.sort_values(ascending=False, inplace=True)
    
    ax = axes[i]
    city_data.plot(kind='bar', ax=ax, color='skyblue')
    ax.set_title(f'Major Complaint Types in {city}')
    ax.set_xlabel('Complaint Type')
    ax.set_ylabel('Count')
    ax.set_xticklabels(city_data.index, rotation=45)
    
plt.tight_layout()[]
plt.show()
```

```python
any['Created Date'] = pd.to_datetime(any['Created Date'])
any['Closed Date'] = pd.to_datetime(any['Closed Date'])
```

```python
any['Response Time'] = (any['Closed Date'] - any['Created Date']).dt.total_seconds()
```

```python
print(any[['Created Date', 'Closed Date', 'Response Time']].head())
```

```python
average_response_time = any.groupby('Complaint Type')['Response Time'].mean().sort_values(ascending=False)
```

```python
print(average_response_time)
```

```python
plt.figure(figsize=(12,6))
average_response_time.plot(kind='bar', color='skyblue')
plt.title('Average Response Time by Complaint Type')
plt.xlabel('Complaint Type')
plt.ylabel('Average Response Time (seconds)')
plt.xticks(rotation=90)
plt.show()
```
